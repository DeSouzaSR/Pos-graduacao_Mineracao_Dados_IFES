## Notas escolares

```python
N = np.random.randn(2000)
N.mean(), N.std()
```

```python
A = 5+2*N
A.mean(), A.std()
```

```python
A[A>10] = 10
A[A<0]  = 0

A
```

```python
sns.histplot(A, bins=10, kde=True)
```

```python
np.histogram(A, bins=10)
```

```python
df.describe()
```

```python
df.boxplot()
```

### Dataframe com dicionario

```python
frutas = {'maca':6, 'morango':4, 'melancia':3, 'uva': 3, 'pessego':1, 'banana':6, 'laranja':7}

df_frutas = pd.DataFrame.from_dict(frutas, orient='index', columns=['Quantidade'])
```

```python
df_frutas
```

```python
total_frutas = df_frutas['Quantidade'].sum()
```

```python
df_frutas['FreqRelat'] = df_frutas['Quantidade'] / total_frutas
```

```python
df_frutas
```

```python
df_frutas.plot.bar(y='Quantidade')
```

## IRIS DATASET

```python
from sklearn.datasets import load_iris
```

```python
data_iris = load_iris()
```

```python
#data_iris
```

```python
print(data_iris.DESCR)
```

```python
X = pd.DataFrame(data_iris['data'], columns=data_iris['feature_names'])
X
```

```python
y=pd.Series(data_iris['target'], name='Target')
y
```

```python
X.describe()
```

```python
df = X
df['target'] = y
```

```python
sns.pairplot(df, height=1.5, hue='target', palette='tab10')
```

```python
X.corr(method='pearson')
```

```python
X = X.drop('target', axis=1)
```
