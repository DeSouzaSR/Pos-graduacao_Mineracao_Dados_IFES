# Exemplo básico de Networkx

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo não direcionado
G = nx.Graph()

# Adicionando arestas com base em participações conjuntas em projetos
# Ex: Aluno1 e Aluno2 participaram juntos do Projeto A
G.add_edges_from([
    ("Ana", "Bruno"),
    ("Ana", "Gustavo"),
    ("Bruno", "Gustavo"),
    ("Gustavo", "Diana"),
    ("Diana", "Eduarda"),
    ("Eduarda", "Felipe"),
    ("Gustavo", "Felipe")
])

# Desenhando o grafo
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightgreen', node_size=1000, edge_color='gray')
plt.title("Rede de Colaboração entre Alunos")
plt.show()
```

```python
# Centralidade de grau: quais alunos estão mais conectados?
graus = nx.degree_centrality(G)
print("Centralidade de Grau:")
for aluno, valor in graus.items():
    print(f"{aluno}: {valor:.2f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar grafo não direcionado
G = nx.Graph()

# Relações de amizade (bidirecional)
G.add_edges_from([
    ("Ana", "Bruno"),
    ("Bruno", "Carlos"),
    ("Carlos", "Diana"),
    ("Diana", "Ana")
])

# Visualização
plt.figure(figsize=(6, 4))
nx.draw(G, with_labels=True, node_color="lightblue", node_size=1000, edge_color="gray")
plt.title("Grafo Não Direcionado (Amizade)")
plt.show()
```

Grafo direcionado

```python
DG = nx.DiGraph()  # DiGraph = grafo direcionado

# Relações de "segue"
DG.add_edges_from([
    ("Ana", "Bruno"),   # Ana segue Bruno
    ("Bruno", "Carlos"),
    ("Carlos", "Diana"),
    ("Diana", "Ana"),
    ("Ana", "Carlos")   # Ana também segue Carlos
])

# Visualização
plt.figure(figsize=(6, 4))
nx.draw(DG, with_labels=True, node_color="lightgreen", node_size=1000,
        edge_color="gray", arrows=True)
plt.title("Grafo Direcionado (Segue no Instagram)")
plt.show()
```

Gráfico não ponderado

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar grafo não ponderado
G1 = nx.Graph()

# Arestas representam apenas que há uma relação (ex: participaram juntos)
G1.add_edges_from([
    ("Ana", "Bruno"),
    ("Ana", "Carlos"),
    ("Carlos", "Diana")
])

# Visualização
plt.figure(figsize=(6, 4))
nx.draw(G1, with_labels=True, node_color="lightblue", node_size=1000, edge_color="gray")
plt.title("Grafo Não Ponderado (relação simples)")
plt.show()
```

Gráfico ponderado

```python
G2 = nx.Graph()

# Arestas com pesos (ex: quantidade de mensagens trocadas)
G2.add_edge("Ana", "Bruno", weight=3)
G2.add_edge("Ana", "Carlos", weight=1)
G2.add_edge("Carlos", "Diana", weight=5)

# Extraindo pesos para exibir
edge_labels = nx.get_edge_attributes(G2, 'weight')

# Visualização com pesos
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G2)
nx.draw(G2, pos, with_labels=True, node_color="lightgreen", node_size=1000, edge_color="gray")
nx.draw_networkx_edge_labels(G2, pos, edge_labels=edge_labels)
plt.title("Grafo Ponderado (com peso nas relações)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo simples
G = nx.Graph()

# Relações de colaboração simples (cada par indica que trabalharam juntos em algum momento)
G.add_edges_from([
    ("Ana", "Bruno"),
    ("Ana", "Carlos"),
    ("Bruno", "Diana"),
    ("Carlos", "Eduarda"),
    ("Diana", "Felipe"),
    ("Felipe", "Gabriel"),
    ("Gabriel", "Ana")
])

# Visualização
plt.figure(figsize=(7, 5))
nx.draw(G, with_labels=True, node_color="lightblue", node_size=1000, edge_color="gray")
plt.title("Grafo Simples")
plt.show()
```

Multigrafo

```python
MG = nx.MultiGraph()

# Vários alunos trabalhando juntos em diferentes projetos
MG.add_edge("Ana", "Bruno", label="Projeto A")
MG.add_edge("Ana", "Bruno", label="Projeto B")
MG.add_edge("Carlos", "Bruno", label="Projeto C")
MG.add_edge("Carlos", "Diana", label="Projeto A")
MG.add_edge("Eduarda", "Diana", label="Projeto B")
MG.add_edge("Eduarda", "Felipe", label="Projeto C")
MG.add_edge("Felipe", "Gabriel", label="Projeto A")
MG.add_edge("Ana", "Felipe", label="Projeto D")
MG.add_edge("Ana", "Felipe", label="Projeto E")

# Visualizar com labels
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(MG, seed=42)
nx.draw(MG, pos, with_labels=True, node_color="lightgreen", node_size=1000, edge_color="gray")

# Exibir rótulos das arestas
edge_labels = {(u, v, k): d['label'] for u, v, k, d in MG.edges(keys=True, data=True)}
nx.draw_networkx_edge_labels(MG, pos, edge_labels=edge_labels, font_size=8)
plt.title("Multigrafo com Alunos e Projetos Diferentes")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar o grafo simples
G = nx.Graph()
G.add_edge("Ana", "Bruno", label="Projeto B")
G.add_edge("Ana", "Carlos", label="Projeto C")
G.add_edge("Carlos", "Diana", label="Projeto Z")
G.add_edge("Diana", "Bruno", label="Projeto E")

# Criar o multigrafo
MG = nx.MultiGraph()
MG.add_edge("Ana", "Bruno", label="Projeto A")
MG.add_edge("Ana", "Bruno", label="Projeto B")
MG.add_edge("Ana", "Carlos", label="Projeto C")
MG.add_edge("Carlos", "Diana", label="Projeto D")
MG.add_edge("Carlos", "Diana", label="Projeto Z")
MG.add_edge("Diana", "Bruno", label="Projeto E")

# Layout fixo
pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(14, 6))

# Grafo Simples
plt.subplot(1, 2, 1)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1200)
edge_labels_G = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels_G, font_size=9)
plt.title("Grafo Simples (1 ligação por par)")

# Multigrafo com curvas para múltiplas arestas
plt.subplot(1, 2, 2)
nx.draw(MG, pos, with_labels=True, node_color='lightgreen', node_size=1200)

# Desenhar múltiplas arestas com pequenas curvas
for i, (u, v, k, d) in enumerate(MG.edges(keys=True, data=True)):
    rad = 0.1 * (k + 1)  # curva variável por chave
    nx.draw_networkx_edges(MG, pos, edgelist=[(u, v)], connectionstyle=f"arc3,rad={rad}", edge_color='gray')
    mid_x = (pos[u][0] + pos[v][0]) / 2
    mid_y = (pos[u][1] + pos[v][1]) / 2 + rad  # desloca o label com a curva
    plt.text(mid_x, mid_y, d['label'], fontsize=8, color='black')

plt.title("Multigrafo (múltiplas arestas visíveis)")
plt.tight_layout()
plt.show()
```

Gráfico bipartido

```python
import networkx as nx
import matplotlib.pyplot as plt

# Grupos
alunos = ["Ana", "Bruno", "Carlos"]
projetos = ["Projeto A", "Projeto B"]

# Criar grafo bipartido
B = nx.Graph()

# Adicionar nós com atributo de grupo (0: aluno, 1: projeto)
B.add_nodes_from(alunos, bipartite=0)
B.add_nodes_from(projetos, bipartite=1)

# Arestas entre alunos e projetos
B.add_edges_from([
    ("Ana", "Projeto A"),
    ("Bruno", "Projeto A"),
    ("Bruno", "Projeto B"),
    ("Carlos", "Projeto B")
])

# Separar grupos para layout
pos = nx.bipartite_layout(B, alunos)

# Desenhar
plt.figure(figsize=(7, 4))
nx.draw(B, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=1200)
plt.title("Grafo Bipartido: Alunos x Projetos")
plt.show()
```

# Metricas com Networkx

```python
import networkx as nx

# Criar grafo
G = nx.Graph()
G.add_edges_from([
    ("Ana", "Bruno"),
    ("Ana", "Carlos"),
    ("Carlos", "Diana"),
    ("Bruno", "Eduarda"),
    ("Eduarda", "Felipe"),
    ("Carlos", "Felipe"),
])

# Calcular métricas
grau = dict(G.degree())
centralidade_betweenness = nx.betweenness_centrality(G)
centralidade_closeness = nx.closeness_centrality(G)
pagerank = nx.pagerank(G)

# Mostrar resultados
for node in G.nodes():
    print(f"Nó: {node}")
    print(f"  Grau: {grau[node]}")
    print(f"  Betweenness: {centralidade_betweenness[node]:.3f}")
    print(f"  Closeness: {centralidade_closeness[node]:.3f}")
    print(f"  PageRank: {pagerank[node]:.3f}")
    print("-" * 30)
```

## visualização com Pyvis

```python
from pyvis.network import Network

net = Network(notebook=True)
for node in G.nodes():
    title = f"""
    Grau: {grau[node]}<br>
    Betweenness: {centralidade_betweenness[node]:.3f}<br>
    Closeness: {centralidade_closeness[node]:.3f}<br>
    PageRank: {pagerank[node]:.3f}
    """
    net.add_node(node, label=node, title=title, value=grau[node])

for edge in G.edges():
    net.add_edge(edge[0], edge[1])

net.show("grafo_metricas.html")
```
