## Exemplos basicos de NLTK

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from wordcloud import WordCloud

import spacy
# !python -m spacy download pt_core_news_sm
```

```python
texto = "A aula foi excelente e muito produtiva"
```

```python
tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(texto)
print(tokens)
```

## Tokenização e Stopwords

```python
nltk.download('stopwords')

texto = "Achei a disciplina muito interessante, mas o material poderia ter mais exemplos."

# Usando TreebankWordTokenizer (não depende do punkt)
tokenizer = TreebankWordTokenizer()
tokens = tokenizer.tokenize(texto)

# Remover stopwords
stop_words = set(stopwords.words('portuguese'))
tokens_filtrados = [word for word in tokens if word.lower() not in stop_words]

print(tokens_filtrados)
```

## Spacy

```python
# Carregar modelo em português
nlp = spacy.load("pt_core_news_sm")

texto = "A aula foi excelente e muito produtiva."

# Processar
doc = nlp(texto)

# Tokens
tokens = [token.text for token in doc]
print(tokens)
```

Exemplo básico 2 no Spacy

```python
# Carregar o modelo de português
nlp = spacy.load("pt_core_news_sm")

texto = "A disciplina de mineração de dados é muito interessante."

# Processar o texto
doc = nlp(texto)

# Exibir tokens
print([token.text for token in doc])

# Exibir lemas
print([token.lemma_ for token in doc])
```

## Transformers

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer("Achei a disciplina muito interessante e bem organizada!")
print(tokens)
```

## Worldcloud

```python
import matplotlib.pyplot as plt

texto = """
A metodologia do curso foi excelente, gostei muito das atividades práticas.
O professor apresentou domínio do conteúdo, mas senti falta de mais exemplos em sala.
Recomendaria este curso para outros colegas.
"""

# Gerar nuvem
wordcloud = WordCloud(width=800, height=400, background_color='white', collocations=False).generate(texto)

# Exibir
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
