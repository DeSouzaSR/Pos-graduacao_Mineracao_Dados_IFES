# Agrupando os dados por região

```python
grupos_regiao = dataframe.groupby('regiao')
grupos_regiao
```

```python
print(len(grupos_regiao))
```

```python
for regiao, dataframe_regiao in grupos_regiao:
    media_nota_redacao = dataframe_regiao['nota_redacao'].mean() # Cálculo da média
    desvio_padrao_nota_redacao = dataframe_regiao['nota_redacao'].std() # Cálculo do desvio padrão
    menor_nota = dataframe_regiao['nota_redacao'].min()
    maior_nota = dataframe_regiao['nota_redacao'].max()
    amplitude = maior_nota - menor_nota

    #print(f'\nRegião: {regiao} {len(dataframe_regiao)}')
    print(regiao)
    print(f'{"Quantidade de dados":<30}{len(dataframe_regiao):>7d}')
    print(f'{"Média Nota Redação":<30} {media_nota_redacao:>6.2f}')
    print(f'{"Desvio Padrão Nota Redação":<30} {desvio_padrao_nota_redacao:>6.2f}')
    print(f'{"Menor Nota Redação":<30} {menor_nota:>6.2f}')
    print(f'{"Maior Nota Redação":<30} {maior_nota:>6.2f}')
    print(f"{'Amplitude':<30} {amplitude:>6.2f}\n")
```
