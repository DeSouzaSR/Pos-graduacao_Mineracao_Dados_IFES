### Histograma

### Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

dados = np.random.normal(170, 10, 250)

plt.hist(dados)

plt.show() # Exibe na tela uma janela com a figura gerada.
```

### Pandas

```python
import pandas as pd

dados = np.random.normal(170, 10, 250)

dataframe = pd.DataFrame({
    'dados': dados
})

dataframe['dados'].hist()

plt.show() # Para quem está executando localmente no seu computador
```

### Seaborn

```python
import seaborn as sns
import pandas as pd

dados = np.random.normal(170, 10, 250)

dataframe = pd.DataFrame({
    'dados': dados
})

sns.histplot(dataframe, x='dados')

plt.savefig('histograma.png', dpi=300)

plt.show() # Para quem está executando localmente no seu computador
```

### Gráfico de Pizza

###Matplotlib

```python
import matplotlib.pyplot as plt

percentuais = [39.55, 60.45]
tipos = ['Pública', 'Privada']

plt.pie(percentuais, labels=tipos, autopct='%1.2f%%')

plt.show()
```

### Pandas

```python
import pandas as pd

percentuais = [39.55, 60.45]
tipos = ['Pública', 'Privada']

dataframe = pd.DataFrame({
    'tipo': tipos,
    'percentual': percentuais
})

dataframe.groupby(['tipo']).sum().plot(kind='pie', y='percentual',
                                       autopct='%1.2f%%')
```

### Prática

```python
import pandas as pd

dataframe = pd.read_csv('/content/igm_modificado.csv', encoding='utf-8')

dataframe
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(dataframe, x='nota_redacao', color='#B0E0E6', edgecolor='blue')

plt.title('Distribuição de Notas da Prova de Redação')
plt.xlabel('Nota')
plt.ylabel('Frequência')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(dataframe, x='nota_mat', color='LightGrey', edgecolor='blue')

plt.title('Distribuição de Notas do Exame de Matemática')
plt.xlabel('Nota')
plt.ylabel('Frequência')

plt.show()
```

```python
sns.boxplot(dataframe, y='nota_ciencias', palette='rainbow', width=0.2)

plt.title('Distribuição das Notas da prova de Ciências')
plt.xlabel('Frequência')
plt.ylabel('Notas')

plt.ylim(350, 600)

plt.show()
```

```python
sns.violinplot(dataframe, y='nota_mat', width=0.3)

plt.title('Distribuição das Notas do Exame de Matemática')
plt.xlabel('Frequência')
plt.ylabel('Notas')

plt.show()
```

```python
from collections import Counter

lista_valores = dataframe['porte'].tolist()

contagem = Counter(lista_valores)

percentuais = []
categorias = []

for tipo, frequencia in contagem.items():
  categorias.append(tipo)
  percentuais.append(frequencia)

soma_frequencia = sum(percentuais)

percentuais = [100 * (p / soma_frequencia) for p in percentuais]

plt.pie(percentuais, labels=categorias, autopct='%.2f%%')

plt.legend(categorias, loc='lower left')

plt.show()
```
