### Média Aritmética

```python
notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0]

media_aritmetica = sum(notas_alunos) / len(notas_alunos)

print(f'Média Aritmética: {media_aritmetica: .2f}')
```

```python
import numpy as np

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0]

media_aritmetica_np = np.mean(notas_alunos)

print(f'Média Aritmética: {media_aritmetica_np: .2f}')
```

```python
import pandas as pd

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0]

dataframe = pd.DataFrame({'Notas': notas_alunos})

media_notas_pd = dataframe['Notas'].mean()

print(f'Média Aritmética: {media_notas_pd: .2f}')
```

### Média Ponderada

```python
notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0]
pesos =        [0.1, 0.2, 0.2, 0.3, 0.2]

media_ponderada = 0

for nota, peso in zip(notas_alunos, pesos):
  media_ponderada += nota * peso

media_poderada = media_ponderada / sum(pesos)

print(f'\nMédia Ponderada: {media_ponderada: .2f}')
```

```python
import numpy as np

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0]
pesos =        [0.1, 0.2, 0.2, 0.3, 0.2]

media_ponderada_np = np.average(notas_alunos, weights=pesos)

print(f'Média Ponderada: {media_ponderada: .2f}')
```

### Moda

```python
# Opção 1

from collections import Counter

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0,
                4.1, 8.5, 8.5]

c = Counter(notas_alunos)

moda = c.most_common()

print(f'Moda: {moda}')
```

```python
# Opção 2 - Pandas

import pandas as pd

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0,
                4.1, 8.5, 8.5]

dataframe = pd.DataFrame({'Notas': notas_alunos})

moda = dataframe['Notas'].mode()

print(f'Moda: {moda}')
```

```python
# Opção 3 - Statistics

import statistics as st

moda = st.mode(notas_alunos) # OBS: A função mode só retorna o elemento mais frequente.

moda = st.multimode(notas_alunos)

print(f'Moda: {moda}')
```

## Amplitude

```python
# Opção 1

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0, 4.1, 8.5]

menor_nota = min(notas_alunos)
maior_nota = max(notas_alunos)

amplitude = maior_nota - menor_nota

print(f'\nMenor Nota: {menor_nota}')
print(f'Maior Nota: {maior_nota}')

print(f'\nAmplitude: {amplitude}')
```

```python
# Opção 2 - Usando Numpy

import numpy as np

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0, 4.1, 8.5]

menor_nota = np.min(notas_alunos)
maior_nota = np.max(notas_alunos)

amplitude = np.ptp(notas_alunos)

print(f'\nMenor Nota: {menor_nota}')
print(f'Maior Nota: {maior_nota}')

print(f'\nAmplitude: {amplitude}')
```

```python
# Opção 3 - Usando Pandas

import pandas as pd

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0, 4.1, 8.5]

dataframe = pd.DataFrame({'Notas': notas_alunos})

menor_nota = dataframe['Notas'].min()
maior_nota = dataframe['Notas'].max()

amplitude = maior_nota - menor_nota

print(f'\nMenor Nota: {menor_nota}')
print(f'Maior Nota: {maior_nota}')

print(f'\nAmplitude: {amplitude}')
```

## Variância e Desvio Padrão

```python
# Opção 1

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0, 4.1, 8.5]

media = sum(notas_alunos) / len(notas_alunos)

variancia = 0

for nota in notas_alunos:
  variancia += (media - nota) ** 2

variancia = variancia / (len(notas_alunos) - 1)

desvio_padrao = variancia ** 0.5

print(f'\nVariância: {variancia: .2f}')

print(f'\nDesvio Padrão: {desvio_padrao: .2f}')
```

```python
# Opção 2 - Numpy

import numpy as np

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0, 4.1, 8.5]

variancia = np.var(notas_alunos, ddof=1)

desvio_padrao = np.std(notas_alunos, ddof=1)

print(f'\nVariância: {variancia: .2f}')

print(f'\nDesvio Padrão: {desvio_padrao: .2f}')
```

```python
# Opção 3 - Pandas

import pandas as pd

notas_alunos = [9.5, 7.2, 4.1, 8.9, 10.0, 4.1, 8.5, 6.3, 9.0, 4.1, 8.5]

dataframe = pd.DataFrame({'Notas': notas_alunos})

variancia = dataframe['Notas'].var(ddof=1)
desvio_padrao = dataframe['Notas'].std(ddof=1)

print(f'\nVariância: {variancia: .2f}')
print(f'\nDesvio Padrão: {desvio_padrao: .2f}')
```

### Prática

```python
import pandas as pd

caminho_csv = '/content/igm_modificado.csv'

dataframe = pd.read_csv(caminho_csv, sep=',', encoding='utf-8', thousands=',')

dataframe
```

```python
dataframe.describe()
```

```python
dataframe_es = dataframe[dataframe['estado'] == 'ES']

dataframe_es
```

```python
dataframe_pb = dataframe[dataframe['estado'] == 'PB']

dataframe_pb
```

```python
grupos_regiao = dataframe.groupby('regiao')

grupos_regiao
```

```python
print(len(grupos_regiao))
```

```python
for regiao, dataframe_regiao in grupos_regiao:

  print(f'\nRegião: {regiao} -- {len(dataframe_regiao)}')

  media_nota_redacao = dataframe_regiao['nota_redacao'].mean() # Cálculo da média
  desvio_padrao_nota_redacao = dataframe_regiao['nota_redacao'].std() # Cálculo do desvio padrão

  menor_nota = dataframe_regiao['nota_redacao'].min()
  maior_nota = dataframe_regiao['nota_redacao'].max()

  amplitude = maior_nota - menor_nota

  print(f'\n\tMédia Nota Redação: {media_nota_redacao: .2f}')
  print(f'\tDesvio Padrão Nota Redação: {desvio_padrao_nota_redacao: .2f}')

  print(f'\n\tMenor Nota Redação: {menor_nota: .2f}')
  print(f'\tMaior Nota Redação: {maior_nota: .2f}')
  print(f'\tAmplitude: {amplitude: .2f}')
```

```python
# | ==> ou
# & ==> e

dataframe_es_pb = dataframe[
    (dataframe['estado'] == 'ES') | (dataframe['estado'] == 'PB')
    ]

dataframe_es_pb
```

```python
dataframe_pb = dataframe[
    ((dataframe['estado'] == 'PB') | (dataframe['estado'] == 'ES')) & (dataframe['municipio'].str.startswith('V'))
    ]

dataframe_pb
```
