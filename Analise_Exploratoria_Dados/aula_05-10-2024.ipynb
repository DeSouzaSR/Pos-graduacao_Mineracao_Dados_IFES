## Gráfico de pizza

```python
lista_valores = df['porte'].tolist()
```

```python
from collections import Counter

valores_distintos = Counter(lista_valores)
```

```python
percentuais = []
categorias = []

for tipo, frequencia in valores_distintos.items():
    percentuais.append(frequencia)
    categorias.append(tipo)
```

```python
soma_frequencia = sum(percentuais)
```

```python
percentuais = [100 * (p / soma_frequencia) for p in percentuais]
```

```python
percentuais
```

```python
plt.pie(percentuais, labels=categorias, autopct="%.2f%%")
plt.show()
```
